````python
import fastf1
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import logging
from datetime import datetime


# ============================================================================
# LOGGING SETUP
# ============================================================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)


# ============================================================================
# CONFIG
# ============================================================================
CACHE_DIR = r'C:\Users\moina\f1_cache'
OUTPUT_FILE = 'f1_training_dataset_2018_2024.csv'
YEARS = [2018, 2019, 2020, 2021, 2022, 2023, 2024]


# Enable FastF1 cache
fastf1.Cache.enable_cache(CACHE_DIR)


# ============================================================================
# UTILITY FUNCTIONS (FINAL CORRECTED VERSION)
# ============================================================================


def get_session_safe(year, event_name):
    """
    Try to get qualifying and race sessions independently.
    FastF1 v3.6.1 column structure requires Abbreviation (not Driver in results).
    
    Args:
        year: F1 season
        event_name: Event name from schedule
    
    Returns:
        tuple: (qualifying_session or None, race_session or None)
    """
    q_types = ['Q', 'SQ', 'S']  # Try Qualifying ‚Üí Sprint Q ‚Üí Sprint
    r_types = ['R']
    
    q_session = None
    r_session = None
    
    # Try qualifying independently
    for q_type in q_types:
        try:
            q_session = fastf1.get_session(year, event_name, q_type)
            logger.debug(f"‚úì Got {q_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {q_type} failed: {type(e).__name__}")
            continue
    
    if q_session is None:
        logger.debug(f"No qualifying session found for {event_name}")
    
    # Try race independently (regardless of qual)
    for r_type in r_types:
        try:
            r_session = fastf1.get_session(year, event_name, r_type)
            logger.debug(f"‚úì Got {r_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {r_type} failed: {type(e).__name__}")
            continue
    
    if r_session is None:
        logger.debug(f"No race session found for {event_name}")
    
    return q_session, r_session


def classify_dnf_corrected(status):
    """
    CORRECTED DNF Classification (verified against 2018 F1 data).
    
    Returns 1 if driver Did Not Finish (not classified in final results)
    Returns 0 if driver completed race and was classified (regardless of laps down)
    Returns 0 if status is NaN (default to finished)
    
    Logic:
    - 'Finished' ‚Üí DNF=0 (classified, completed lead lap)
    - '+1 Lap', '+2 Laps', etc. ‚Üí DNF=0 (classified, lapped but finished)
    - 'Engine', 'Power loss', 'Collision', etc. ‚Üí DNF=1 (retired, not classified)
    
    Args:
        status: Status value from FastF1 results
    
    Returns:
        0 if finished/classified, 1 if DNF/not classified, NaN if unknown
    """
    if pd.isna(status):
        return 0
    
    status_str = str(status).strip()
    status_lower = status_str.lower()
    
    # CLASSIFIED (Finished) - Return 0
    # Check for '+X Lap(s)' pattern (e.g., '+1 Lap', '+2 Laps')
    if '+' in status_lower and 'lap' in status_lower:
        return 0
    
    # Check for exact match with 'Finished'
    if status_lower == 'finished':
        return 0
    
    # NOT CLASSIFIED (DNF) - Return 1
    # Everything else is a retirement/non-classification
    return 1


def get_event_date_safe(event_row, schedule_idx):
    """
    Extract event date with multiple fallbacks.
    
    Args:
        event_row: Row from FastF1 schedule
        schedule_idx: Position in schedule
    
    Returns:
        datetime or NaT
    """
    date_columns = ['EventDate', 'Session5Date', 'RaceDate', 'Date']
    
    for col in date_columns:
        if col in event_row.index:
            date_val = event_row.get(col, pd.NaT)
            if pd.notna(date_val):
                return pd.to_datetime(date_val, errors='coerce')
    
    logger.debug(f"No valid date found in schedule row {schedule_idx}")
    return pd.NaT


def get_team_perf_score(df_race_results):
    """
    Calculate team performance from actual race results.
    
    Args:
        df_race_results: DataFrame with 'team' and 'finishing_position' columns
    
    Returns:
        dict: team -> performance_score (0-1)
    """
    team_finishes = df_race_results[df_race_results['finishing_position'].notna()].groupby('team').size()
    if len(team_finishes) == 0:
        return {}
    max_finishes = team_finishes.max()
    team_perf = (team_finishes / max_finishes).to_dict()
    return team_perf


def calculate_elo_safe(results_df, k=24, initial_elo=1500):
    """
    Calculate Elo ratings with robust schedule_idx handling.
    
    Args:
        results_df: DataFrame with race results
        k: Elo K-factor (default 24)
        initial_elo: Starting Elo rating (default 1500)
    
    Returns:
        tuple: (elo_history_df, final_elo_df)
    """
    res = results_df.copy()
    res["finishing_position"] = pd.to_numeric(res["finishing_position"], errors="coerce")
    res = res[res["finishing_position"].notna()].copy()

    if len(res) == 0:
        logger.warning("No finished races found for Elo calculation")
        return pd.DataFrame(), pd.DataFrame()

    # Generate per-year ordering if schedule_idx missing
    if 'schedule_idx' not in res.columns or res['schedule_idx'].isna().all():
        logger.warning("schedule_idx missing ‚Äî generating per-year ordering")
        res['schedule_idx'] = res.groupby('race_year').cumcount()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    else:
        missing_idx = res['schedule_idx'].isna().sum()
        if missing_idx > 0:
            logger.warning(f"schedule_idx has {missing_idx} NaN values ‚Äî filling with per-year rank")
            res.loc[res['schedule_idx'].isna(), 'schedule_idx'] = \
                res[res['schedule_idx'].isna()].groupby('race_year').cumcount() + res['schedule_idx'].max()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    
    # Build race_order
    race_order = res[["race_year", "race_key", "event_date", "schedule_idx"]].drop_duplicates()
    race_order = race_order.sort_values(sort_keys).reset_index(drop=True)
    
    order_map = {rk: i for i, rk in enumerate(race_order["race_key"].tolist())}
    res["race_order_idx"] = res["race_key"].map(order_map)
    
    # Fail loudly if mapping failed
    unmapped = res[res["race_order_idx"].isna()]["race_key"].unique()
    if len(unmapped) > 0:
        raise ValueError(f"‚ùå CRITICAL: These race_keys were not mapped to order: {unmapped}")
    
    res = res.sort_values(["race_order_idx", "finishing_position"]).reset_index(drop=True)

    elo = {}
    history = []

    for rk in res["race_key"].unique():
        race_rows = res[res["race_key"] == rk].sort_values("finishing_position")
        drivers_in_race = race_rows["driver"].tolist()
        positions = race_rows["finishing_position"].astype(int).tolist()
        n = len(drivers_in_race)

        if n <= 0:
            continue

        for d in drivers_in_race:
            elo.setdefault(d, initial_elo)

        deltas = {}
        for i, d_i in enumerate(drivers_in_race):
            elo_i = elo[d_i]
            exp_sum = 0.0
            for j, d_j in enumerate(drivers_in_race):
                if d_i == d_j:
                    continue
                diff = elo[d_j] - elo_i
                p = 1.0 / (1.0 + 10 ** (diff / 400.0))
                exp_sum += p
            expected_norm = exp_sum / (n - 1) if n > 1 else 0.0
            actual_norm = (n - positions[i]) / (n - 1) if n > 1 else 0.0
            delta = k * (actual_norm - expected_norm)
            deltas[d_i] = delta

        for d in drivers_in_race:
            history.append({
                "race_key": rk,
                "driver": d,
                "elo_before": elo[d],
                "elo_delta": deltas.get(d, 0)
            })

        for d, delta in deltas.items():
            elo[d] = elo[d] + delta

    final_elo = pd.DataFrame([(d, e) for d, e in elo.items()], columns=['driver', 'elo_final'])
    return pd.DataFrame(history), final_elo


def calculate_overtaking_difficulty_from_data(df):
    """Calculate overtaking difficulty based on average position changes per circuit"""
    if 'position_changes' not in df.columns:
        logger.warning("position_changes column not found.")
        return {}
    
    circuit_overtaking = df.groupby('circuit').agg({
        'position_changes': 'mean'
    }).dropna()
    
    if len(circuit_overtaking) == 0:
        logger.warning("No valid position_changes data for circuits.")
        return {}
    
    max_changes = circuit_overtaking['position_changes'].max()
    if max_changes > 0:
        circuit_overtaking['difficulty'] = 1 - (circuit_overtaking['position_changes'] / max_changes)
    else:
        circuit_overtaking['difficulty'] = 0.6
    
    return circuit_overtaking['difficulty'].to_dict()


# ============================================================================
# MAIN DATA EXTRACTION
# ============================================================================

logger.info(f"Starting F1 data extraction: {YEARS}")
all_rows = []
successful_years = []
failed_years = []

for year in YEARS:
    try:
        logger.info(f"üìÖ Processing season {year}")
        schedule = fastf1.get_event_schedule(year, include_testing=False)
        logger.info(f"Found {len(schedule)} events")
        
        year_rows = 0
        for schedule_idx, (event_idx, event_row) in enumerate(schedule.iterrows()):
            event_name = event_row['EventName']
            
            try:
                q, r = get_session_safe(year, event_name)
                
                if q is None and r is None:
                    logger.warning(f"‚è≠Ô∏è Skipping {event_name}: No Q or R session")
                    continue
                
                if q is not None:
                    try:
                        q.load(laps=True, telemetry=False, weather=False)
                    except Exception as e:
                        logger.warning(f"Q load failed: {type(e).__name__}")
                        q = None
                
                if r is not None:
                    try:
                        r.load(laps=True, telemetry=False, weather=False)
                    except Exception as e:
                        logger.warning(f"R load failed: {type(e).__name__}")
                        r = None
                
                if q is None and r is None:
                    logger.warning(f"‚è≠Ô∏è Skipping {event_name}: Failed to load Q and R")
                    continue
                
                q_extracted = pd.DataFrame()
                if q is not None and hasattr(q, 'results') and len(q.results) > 0:
                    q_table = q.results[['Abbreviation', 'Position', 'Q1', 'Q2', 'Q3']].copy()
                    q_table.columns = ['driver', 'qualifying_position', 'q1', 'q2', 'q3']
                    q_extracted = q_table[['driver', 'qualifying_position']].copy()
                    q_extracted['qualifying_position'] = pd.to_numeric(q_extracted['qualifying_position'], errors='coerce')
                
                r_extracted = pd.DataFrame()
                if r is not None and hasattr(r, 'results') and len(r.results) > 0:
                    r_table = r.results[['Abbreviation', 'Position', 'Status', 'Points', 'Laps']].copy()
                    r_table.columns = ['driver', 'finishing_position', 'status', 'points', 'laps_completed']
                    r_extracted = r_table[['driver', 'finishing_position', 'status', 'points', 'laps_completed']].copy()
                    r_extracted['finishing_position'] = pd.to_numeric(r_extracted['finishing_position'], errors='coerce')
                
                if len(r_extracted) == 0:
                    logger.warning(f"‚è≠Ô∏è Skipping {event_name}: No race results")
                    continue
                
                merged = r_extracted.copy()
                if len(q_extracted) > 0:
                    merged = merged.merge(q_extracted, on='driver', how='left')
                else:
                    merged['qualifying_position'] = np.nan
                
                event_date = get_event_date_safe(event_row, schedule_idx)
                circuit_name = event_row.get('Location', event_name)
                
                for _, row in merged.iterrows():
                    driver = row['driver']
                    dnf = classify_dnf_corrected(row['status'])
                    
                    all_rows.append({
                        'race_key': f"{year}__{schedule_idx:02d}__{event_name.replace(' ', '_')}",
                        'race_year': year,
                        'schedule_idx': schedule_idx,
                        'event': event_name,
                        'circuit': circuit_name,
                        'event_date': event_date,
                        'driver': driver,
                        'qualifying_position': row.get('qualifying_position', np.nan),
                        'race_lap_time_s': np.nan,
                        'finishing_position': row.get('finishing_position', np.nan),
                        'position_changes': np.nan,
                        'dnf': dnf,
                        'dnf_known': True,
                        'status': row.get('status', ''),
                        'points': row.get('points', 0.0),
                        'laps_completed': row.get('laps_completed', np.nan)
                    })
                    year_rows += 1
                
                logger.info(f"  ‚úì {event_name}: extracted {len(merged)} drivers")
            
            except Exception as e:
                logger.error(f"  ‚úó {event_name}: {type(e).__name__}: {str(e)[:100]}")
                continue
        
        logger.info(f"‚úì Year {year}: {year_rows} total rows")
        successful_years.append(year)
    
    except Exception as e:
        logger.error(f"‚úó Year {year} failed: {type(e).__name__}: {str(e)[:100]}")
        failed_years.append(year)
        continue

logger.info(f"\n‚úÖ Extraction complete:")
logger.info(f"  Successful years: {successful_years}")
logger.info(f"  Failed years: {failed_years}")
logger.info(f"  Total rows: {len(all_rows)}")

# ============================================================================
# BUILD DATAFRAME & TEAM INFO
# ============================================================================

df_raw = pd.DataFrame(all_rows)
logger.info(f"DataFrame shape: {df_raw.shape}")

# Extract team from driver (advanced logic if needed)
team_map = {}
for year in successful_years:
    try:
        schedule = fastf1.get_event_schedule(year, include_testing=False)
        for _, event in schedule.iterrows():
            try:
                q, r = get_session_safe(year, event['EventName'])
                if r is not None:
                    r.load(laps=False, telemetry=False, weather=False)
                    if hasattr(r, 'results') and len(r.results) > 0:
                        for _, row in r.results[['Abbreviation', 'Team']].iterrows():
                            team_map[row['Abbreviation']] = row['Team']
            except:
                pass
    except:
        pass

df_raw['team'] = df_raw['driver'].map(team_map).fillna('Unknown')

print("\n" + "="*80)
print("RAW DATASET PREVIEW:")
print("="*80)
print(df_raw.head())
print("\nColumn Types:")
print(df_raw.dtypes)
print("\nMissing Values:")
print(df_raw.isnull().sum())

# ============================================================================
# CALCULATE ELO RATINGS
# ============================================================================

logger.info("üìä Calculating Elo ratings...")
try:
    df_elo_history, df_elo_final = calculate_elo_safe(df_raw)
    logger.info(f"‚úÖ Elo ratings calculated: {len(df_elo_final)} drivers")
except ValueError as e:
    logger.error(f"Elo calculation error: {str(e)}")
    df_elo_history, df_elo_final = pd.DataFrame(), pd.DataFrame()
````

````python
import fastf1
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import logging
from datetime import datetime


# ============================================================================
# LOGGING SETUP
# ============================================================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)


# =================================================================================================================================
# CONFIG
# ============================================================================
CACHE_DIR = r'C:\Users\moina\f1_cache'
OUTPUT_FILE = 'f1_training_dataset_2018_2024.csv'
YEARS = [2018, 2019, 2020, 2021, 2022, 2023, 2024]


# Enable FastF1 cache
fastf1.Cache.enable_cache(CACHE_DIR)


# ============================================================================
# UTILITY FUNCTIONS (FINAL CORRECTED VERSION - COLUMN NAMES FIXED)
# ============================================================================


def get_col(df, possible_names, default=None):
    """
    Safely get column name with fallback names.
    Handles FastF1 column name variations across versions.
    
    Args:
        df: DataFrame or results object with .columns attribute
        possible_names: List of possible column names to try
        default: Default value if no column found
    
    Returns:
        Column name if found, else default
    """
    if not hasattr(df, 'columns'):
        return default
    
    for name in possible_names:
        if name in df.columns:
            return name
    
    return default


def get_session_safe(year, event_name):
    """
    Try to get qualifying and race sessions independently.
    FastF1 v3.6.1 column structure requires Abbreviation (not Driver in results).
    
    Args:
        year: F1 season
        event_name: Event name from schedule
    
    Returns:
        tuple: (qualifying_session or None, race_session or None)
    """
    q_types = ['Q', 'SQ', 'S']  # Try Qualifying ‚Üí Sprint Q ‚Üí Sprint
    r_types = ['R']
    
    q_session = None
    r_session = None
    
    # Try qualifying independently
    for q_type in q_types:
        try:
            q_session = fastf1.get_session(year, event_name, q_type)
            logger.debug(f"‚úì Got {q_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {q_type} failed: {type(e).__name__}")
            continue
    
    if q_session is None:
        logger.debug(f"No qualifying session found for {event_name}")
    
    # Try race independently (regardless of qual)
    for r_type in r_types:
        try:
            r_session = fastf1.get_session(year, event_name, r_type)
            logger.debug(f"‚úì Got {r_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {r_type} failed: {type(e).__name__}")
            continue
    
    if r_session is None:
        logger.debug(f"No race session found for {event_name}")
    
    return q_session, r_session


def classify_dnf_corrected(status):
    """
    CORRECTED DNF Classification (verified against 2018 F1 data).
    
    Returns 1 if driver Did Not Finish (not classified in final results)
    Returns 0 if driver completed race and was classified (regardless of laps down)
    Returns np.nan if status is unknown/ambiguous
    
    Logic:
    - 'Finished' ‚Üí DNF=0 (classified, completed lead lap)
    - '+1 Lap', '+2 Laps', etc. ‚Üí DNF=0 (classified, lapped but finished)
    - 'Engine', 'Power loss', 'Collision', etc. ‚Üí DNF=1 (retired, not classified)
    - Unknown/ambiguous ‚Üí NaN (don't assume)
    
    Args:
        status: Status value from FastF1 results
    
    Returns:
        0 if finished/classified, 1 if DNF/not classified, NaN if unknown
    """
    if pd.isna(status):
        return np.nan  # Changed from 0 - unknown should be NaN
    
    status_str = str(status).strip().lower()
    
    # CLASSIFIED DRIVERS (DNF = 0)
    # 1. Drivers who finished on lead lap
    if status_str == 'finished':
        return 0
    
    # 2. Drivers who finished but were lapped (still classified)
    if '+' in status_str and 'lap' in status_str:
        # Matches: '+1 lap', '+2 laps', '+3 lap', etc.
        # KEY FIX: '+1 Lap' is CLASSIFIED, not DNF
        return 0
    
    # NOT CLASSIFIED DRIVERS (DNF = 1)
    dnf_keywords = ['engine', 'collision', 'accident', 'gearbox', 'hydraulics', 
                    'retired', 'spun off', 'damage', 'electronics', 'brakes',
                    'power loss', 'fuel system', 'transmission', 'suspension']
    if any(kw in status_str for kw in dnf_keywords):
        return 1
    
    # Unknown/ambiguous - return NaN instead of assuming finished
    return np.nan


def get_event_date_safe(event_row, schedule_idx):
    """
    Extract event date with multiple fallbacks.
    
    Args:
        event_row: Row from FastF1 schedule
        schedule_idx: Position in schedule
    
    Returns:
        datetime or NaT
    """
    date_columns = ['EventDate', 'Session5Date', 'RaceDate', 'Date']
    
    for col in date_columns:
        if col in event_row.index:
            date_val = event_row.get(col, pd.NaT)
            if pd.notna(date_val):
                return pd.to_datetime(date_val, errors='coerce')
    
    logger.debug(f"No valid date found in schedule row {schedule_idx}")
    return pd.NaT


def get_team_perf_score(df_race_results):
    """
    Calculate team performance from actual race results.
    
    Args:
        df_race_results: DataFrame with 'team' and 'finishing_position' columns
    
    Returns:
        dict: team -> performance_score (0-1)
    """
    team_finishes = df_race_results[df_race_results['finishing_position'].notna()].groupby('team').size()
    if len(team_finishes) == 0:
        return {}
    max_finishes = team_finishes.max()
    team_perf = (team_finishes / max_finishes).to_dict()
    return team_perf


def calculate_elo_safe(results_df, k=24, initial_elo=1500):
    """
    Calculate Elo ratings with robust schedule_idx handling.
    
    FIXED: Uses 'finishing_position' column name (matches actual data)
    
    Args:
        results_df: DataFrame with race results
        k: Elo K-factor (default 24)
        initial_elo: Starting Elo rating (default 1500)
    
    Returns:
        tuple: (elo_history_df, final_elo_df)
    """
    res = results_df.copy()
    
    # FIXED: Use correct column name 'finishing_position'
    res["finishing_position"] = pd.to_numeric(res["finishing_position"], errors="coerce")
    res = res[res["finishing_position"].notna()].copy()

    if len(res) == 0:
        logger.warning("No finished races found for Elo calculation")
        return pd.DataFrame(), pd.DataFrame()

    # Generate per-year ordering if schedule_idx missing
    if 'schedule_idx' not in res.columns or res['schedule_idx'].isna().all():
        logger.warning("schedule_idx missing ‚Äî generating per-year ordering")
        res['schedule_idx'] = res.groupby('race_year').cumcount()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    else:
        missing_idx = res['schedule_idx'].isna().sum()
        if missing_idx > 0:
            logger.warning(f"schedule_idx has {missing_idx} NaN values ‚Äî filling with per-year rank")
            res.loc[res['schedule_idx'].isna(), 'schedule_idx'] = \
                res[res['schedule_idx'].isna()].groupby('race_year').cumcount() + res['schedule_idx'].max()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    
    # Build race_order
    race_order = res[["race_year", "race_key", "event_date", "schedule_idx"]].drop_duplicates()
    race_order = race_order.sort_values(sort_keys).reset_index(drop=True)
    
    order_map = {rk: i for i, rk in enumerate(race_order["race_key"].tolist())}
    res["race_order_idx"] = res["race_key"].map(order_map)
    
    # Fail loudly if mapping failed
    unmapped = res[res["race_order_idx"].isna()]["race_key"].unique()
    if len(unmapped) > 0:
        raise ValueError(f"‚ùå CRITICAL: These race_keys were not mapped to order: {unmapped}")
    
    # FIXED: Use correct column name 'finishing_position'
    res = res.sort_values(["race_order_idx", "finishing_position"]).reset_index(drop=True)

    elo = {}
    history = []

    for rk in res["race_key"].unique():
        race_rows = res[res["race_key"] == rk].sort_values("finishing_position")
        drivers_in_race = race_rows["driver"].tolist()
        positions = race_rows["finishing_position"].astype(int).tolist()
        n = len(drivers_in_race)

        if n <= 0:
            continue

        for d in drivers_in_race:
            elo.setdefault(d, initial_elo)

        deltas = {}
        for i, d_i in enumerate(drivers_in_race):
            elo_i = elo[d_i]
            exp_sum = 0.0
            for j, d_j in enumerate(drivers_in_race):
                if d_i == d_j:
                    continue
                diff = elo[d_j] - elo_i
                p = 1.0 / (1.0 + 10 ** (diff / 400.0))
                exp_sum += p
            expected_norm = exp_sum / (n - 1) if n > 1 else 0.0
            actual_norm = (n - positions[i]) / (n - 1) if n > 1 else 0.0
            delta = k * (actual_norm - expected_norm)
            deltas[d_i] = delta

        for d in drivers_in_race:
            history.append({
                "race_key": rk,
                "driver": d,
                "elo_before": elo[d],
                "elo_delta": deltas.get(d, 0)
            })

        for d, delta in deltas.items():
            elo[d] = elo[d] + delta

    final_elo = pd.DataFrame([(d, e) for d, e in elo.items()], columns=['driver', 'elo_final'])
    return pd.DataFrame(history), final_elo


def calculate_overtaking_difficulty_from_data(df):
    """
    Calculate overtaking difficulty based on average position changes per circuit
    
    FIXED: Uses 'position_changes' column name (matches actual data)
    """
    # FIXED: Use correct column name 'position_changes'
    if 'position_changes' not in df.columns:
        logger.warning("position_changes column not found.")
        return {}
    
    circuit_overtaking = df.groupby('circuit').agg({
        'position_changes': 'mean'
    }).dropna()
    
    if len(circuit_overtaking) == 0:
        logger.warning("No valid position_changes data for circuits.")
        return {}
    
    max_changes = circuit_overtaking['position_changes'].max()
    if max_changes > 0:
        circuit_overtaking['difficulty'] = 1 - (circuit_overtaking['position_changes'] / max_changes)
    else:
        circuit_overtaking['difficulty'] = 0.6
    
    return circuit_overtaking['difficulty'].to_dict()


# ============================================================================
# LOAD DATA FROM LOCAL CACHE
# ============================================================================

logger.info("="*80)
logger.info("LOADING F1 DATA FROM LOCAL CACHE - 2018-2024")
logger.info("="*80)
logger.info(f"Cache directory: {CACHE_DIR}\n")

all_rows = []
successful_years = []
failed_years = []

for year in YEARS:
    logger.info(f"\nüìÖ Processing season {year}")
    logger.info("-"*80)
    
    try:
        schedule = fastf1.get_event_schedule(year)
        logger.info(f"Found {len(schedule)} events for {year}")
    except Exception as e:
        logger.error(f"Failed to load schedule: {e}")
        failed_years.append((year, str(e)))
        continue
    
    year_rows = 0
    
    for schedule_idx, (_, event_row) in enumerate(schedule.iterrows()):
        try:
            event_name = event_row.get('EventName') or event_row.get('Event') or 'Unknown'
            circuit_name = event_row.get('OfficialEventName') or event_name
            event_date = get_event_date_safe(event_row, schedule_idx)
            
            race_key = f"{year}__{schedule_idx:02d}__{event_name}".replace(" ", "_").replace("/", "_")
            
            # Get sessions independently
            q, r = get_session_safe(year, event_name)
            
            # Skip only if BOTH missing
            if q is None and r is None:
                logger.debug(f"Skipping {event_name}: no sessions found")
                continue
            
            # Load each session that exists
            if q is not None:
                try:
                    q.load(laps=True, telemetry=False, weather=False)
                except Exception as q_load_err:
                    logger.debug(f"Failed to load qualifying for {event_name}: {q_load_err}")
                    q = None
            
            if r is not None:
                try:
                    r.load(laps=True, telemetry=False, weather=False)
                except Exception as r_load_err:
                    logger.debug(f"Failed to load race for {event_name}: {r_load_err}")
                    r = None
            
            # Skip if BOTH ended up None after load attempts
            if q is None and r is None:
                logger.debug(f"Skipping {event_name}: failed to load either session")
                continue
            
            # Extract qualifying data (using defensive column names)
            q_data = {}
            q_lap_times = {}
            
            if q is not None:
                if q.results is not None and not q.results.empty:
                    try:
                        # Get driver column name with fallback
                        driver_col = get_col(q.results, ['Abbreviation', 'Driver', 'DriverNumber'])
                        team_col = get_col(q.results, ['TeamName', 'Team', 'Constructor'])
                        pos_col = get_col(q.results, ['Position', 'GridPosition', 'P'])
                        status_col = get_col(q.results, ['Status', 'Retired'])
                        
                        if driver_col:
                            for idx, row in q.results.iterrows():
                                driver = row[driver_col]
                                q_data[driver] = {
                                    'team': row.get(team_col, np.nan) if team_col else np.nan,
                                    'qualifying_position': row.get(pos_col, np.nan) if pos_col else np.nan,
                                    'q_status': row.get(status_col, np.nan) if status_col else np.nan
                                }
                        else:
                            logger.debug(f"Failed to find driver column in qualifying results for {event_name}")
                    except Exception as q_extract_err:
                        logger.debug(f"Failed to extract qualifying data for {event_name}: {q_extract_err}")
                
                # Extract qualifying lap times with type guard
                try:
                    if not q.laps.empty:
                        driver_col_laps = get_col(q.laps, ['Driver', 'Abbreviation', 'DriverNumber'])
                        laptime_col = get_col(q.laps, ['LapTime', 'Time'])
                        if driver_col_laps and laptime_col:
                            lap_times_td = q.laps.groupby(driver_col_laps)[laptime_col].mean()
                            # CRITICAL: Check if result is timedelta before calling .dt
                            # Prevents AttributeError if all laps are NaT or column contains strings
                            if pd.api.types.is_timedelta64_dtype(lap_times_td):
                                q_lap_times = lap_times_td.dt.total_seconds().to_dict()
                            else:
                                logger.debug(f"Qualifying lap times not timedelta type for {event_name} (likely red flag session)")
                                q_lap_times = {}
                except Exception as q_laps_err:
                    logger.debug(f"Failed to extract qualifying lap times: {q_laps_err}")
                    q_lap_times = {}
            
            # Extract race data (using defensive column names)
            r_data = {}
            r_lap_times = {}
            
            if r is not None:
                if r.results is not None and not r.results.empty:
                    try:
                        # Get driver column name with fallback
                        driver_col = get_col(r.results, ['Abbreviation', 'Driver', 'DriverNumber'])
                        pos_col = get_col(r.results, ['Position', 'GridPosition', 'P'])
                        status_col = get_col(r.results, ['Status', 'Retired'])
                        points_col = get_col(r.results, ['Points', 'Pts', 'Points'])
                        laps_col = get_col(r.results, ['Laps', 'Lap', 'TotalLaps'])
                        
                        if driver_col:
                            for idx, row in r.results.iterrows():
                                driver = row[driver_col]
                                r_data[driver] = {
                                    'finishing_position': row.get(pos_col, np.nan) if pos_col else np.nan,
                                    'r_status': row.get(status_col, np.nan) if status_col else np.nan,
                                    'points': row.get(points_col, np.nan) if points_col else np.nan,
                                    'laps_completed': row.get(laps_col, np.nan) if laps_col else np.nan
                                }
                        else:
                            logger.debug(f"Failed to find driver column in race results for {event_name}")
                    except Exception as r_extract_err:
                        logger.debug(f"Failed to extract race data for {event_name}: {r_extract_err}")
                
                # Extract race lap times with type guard
                try:
                    if not r.laps.empty:
                        driver_col_laps = get_col(r.laps, ['Driver', 'Abbreviation', 'DriverNumber'])
                        laptime_col = get_col(r.laps, ['LapTime', 'Time'])
                        if driver_col_laps and laptime_col:
                            lap_times_td = r.laps.groupby(driver_col_laps)[laptime_col].mean()
                            # CRITICAL: Check if result is timedelta before calling .dt
                            # Prevents AttributeError if all laps are NaT or column contains strings
                            if pd.api.types.is_timedelta64_dtype(lap_times_td):
                                r_lap_times = lap_times_td.dt.total_seconds().to_dict()
                            else:
                                logger.debug(f"Race lap times not timedelta type for {event_name} (likely red flag session)")
                                r_lap_times = {}
                except Exception as r_laps_err:
                    logger.debug(f"Failed to extract race lap times: {r_laps_err}")
                    r_lap_times = {}
            
            # Skip if no data
            if not q_data and not r_data:
                logger.debug(f"No data extracted for {event_name}")
                continue
            
            # Combine all drivers
            all_drivers = set(q_data.keys()).union(r_data.keys())
            
            for driver in all_drivers:
                q_info = q_data.get(driver, {})
                r_info = r_data.get(driver, {})
                
                qual_pos = q_info.get('qualifying_position', np.nan)
                fin_pos = r_info.get('finishing_position', np.nan)
                r_status = r_info.get('r_status', np.nan)
                
                # CORRECTED DNF classification
                dnf_flag = classify_dnf_corrected(r_status)
                dnf_known = pd.notna(dnf_flag)
                
                # Calculate position changes
                if pd.notna(qual_pos) and pd.notna(fin_pos):
                    position_changes = abs(fin_pos - qual_pos)
                else:
                    position_changes = np.nan
                
                row = {
                    "race_key": race_key,
                    "race_year": year,
                    "schedule_idx": schedule_idx,
                    "event": event_name,
                    "circuit": circuit_name,
                    "event_date": event_date,
                    "driver": driver,  # Abbreviation
                    "team": q_info.get('team', np.nan),
                    "qualifying_position": qual_pos,
                    "qualifying_lap_time_s": q_lap_times.get(driver, np.nan),
                    "race_lap_time_s": r_lap_times.get(driver, np.nan),
                    "finishing_position": fin_pos,
                    "position_changes": position_changes,
                    "dnf": dnf_flag,
                    "dnf_known": dnf_known,
                    "status": r_status,
                    "points": r_info.get('points', np.nan),
                    "laps_completed": r_info.get('laps_completed', np.nan)
                }
                all_rows.append(row)
                year_rows += 1
        
        except Exception as e:
            logger.error(f"Unexpected error processing {event_row.get('EventName', 'Unknown')}: {e}")
            continue
    
    if year_rows > 0:
        successful_years.append((year, year_rows))
        logger.info(f"‚úì Loaded {year_rows} rows from {year}")
    else:
        logger.info(f"‚úó Year {year} produced no rows")

logger.info(f"\n‚úÖ Extraction complete:")
logger.info(f"  Successful years: {successful_years}")
logger.info(f"  Total rows collected: {len(all_rows)}")

# ============================================================================
# CONVERT TO DATAFRAME & ENFORCE SCHEMA
# ============================================================================

if len(all_rows) == 0:
    logger.error("‚ùå No data loaded! Cannot proceed.")
else:
    logger.info(f"\nüìä Converting {len(all_rows)} rows to DataFrame...")
    df_raw = pd.DataFrame(all_rows)
    
    logger.info(f"\n{'='*80}")
    logger.info("SCHEMA VALIDATION & DTYPE ENFORCEMENT")
    logger.info(f"{'='*80}")
    
    # Define expected schema
    schema = {
        'race_year': 'int64',
        'schedule_idx': 'Int64',  # Nullable int
        'qualifying_position': 'float64',
        'finishing_position': 'float64',
        'qualifying_lap_time_s': 'float64',
        'race_lap_time_s': 'float64',
        'position_changes': 'float64',
        'dnf': 'float64',  # Allow NaN now
        'dnf_known': 'bool',
        'points': 'float64',
        'laps_completed': 'float64'
    }
    
    # Enforce dtypes
    logger.info("\nüîç Enforcing column dtypes...")
    for col, dtype in schema.items():
        if col in df_raw.columns:
            try:
                df_raw[col] = df_raw[col].astype(dtype)
                logger.debug(f"  ‚úì {col:30s} ‚Üí {dtype}")
            except Exception as e:
                logger.error(f"  ‚úó Failed to convert {col} to {dtype}: {e}")
                logger.error(f"    Sample values: {df_raw[col].head().tolist()}")
        else:
            logger.warning(f"  ‚ö† Column {col} not found in DataFrame")
    
    # Validate critical columns
    logger.info("\n‚úì Validating critical columns...")
    try:
        assert df_raw['race_year'].notna().all(), "Missing race_year values"
        logger.info("  ‚úì race_year: no missing values")
        
        assert df_raw['driver'].notna().all(), "Missing driver values"
        logger.info("  ‚úì driver: no missing values")
        
        assert df_raw['dnf'].isna().sum() + df_raw['dnf'].isin([0, 1, np.nan]).sum() == len(df_raw), "Invalid dnf values (must be 0, 1, or NaN)"
        logger.info("  ‚úì dnf: all values are 0, 1, or NaN")
        
        logger.info(f"\n‚úÖ Schema validation PASSED")
    except AssertionError as e:
        logger.error(f"\n‚ùå Schema validation FAILED: {e}")
        raise
    
    logger.info(f"\n{'='*80}")
    logger.info("üìä DATASET SUMMARY AFTER VALIDATION")
    logger.info(f"{'='*80}")
    logger.info(f"Shape: {df_raw.shape}")
    logger.info(f"Years: {sorted(df_raw['race_year'].unique())}")
    logger.info(f"Drivers: {df_raw['driver'].nunique()}")
    logger.info(f"Circuits: {df_raw['circuit'].nunique()}")
    logger.info(f"DNF=0 (Finished): {(df_raw['dnf'] == 0).sum()}")
    logger.info(f"DNF=1 (Not Classified): {(df_raw['dnf'] == 1).sum()}")
    logger.info(f"DNF=NaN (Unknown): {df_raw['dnf'].isna().sum()}")
    logger.info(f"{'='*80}\n")
    
    finish_rate = (df_raw['dnf'] == 0).sum() / len(df_raw) * 100 if len(df_raw) > 0 else 0
    print(f"\n‚úÖ Pipeline complete! {len(df_raw)} rows with validated schema")
    print(f"üìä Finish Rate: {finish_rate:.1f}% (should be ~70-80% for typical F1 season)")
````

````python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, f1_score
import logging
import warnings

warnings.filterwarnings('ignore')

# ============================================================================
# LOGGING SETUP
# ============================================================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# ============================================================================
# CONFIG
# ============================================================================
ENGINEERED_DATA = 'f1_features_engineered_2018_2024.csv'
MODEL_OUTPUT = 'f1_xgboost_models_no_leakage'

# ============================================================================
# LOAD DATA
# ============================================================================

logger.info("="*80)
logger.info("XGBoost F1 Model Training (NO DATA LEAKAGE)")
logger.info("="*80)

logger.info(f"\nüìä Loading data from {ENGINEERED_DATA}...")
df = pd.read_csv(ENGINEERED_DATA)
df['event_date'] = pd.to_datetime(df['event_date'])

logger.info(f"‚úì Loaded {len(df)} rows")

# ============================================================================
# REMOVE ROWS WITH NaN TARGET
# ============================================================================

logger.info("\nüßπ Removing rows with NaN target variables...")

df = df[df['finishing_position'].notna()].copy()
df['dnf'] = df['dnf'].fillna(0).astype(int)
df['position_changes'] = df['position_changes'].fillna(0)

logger.info(f"‚úì {len(df)} rows after cleaning")

# ============================================================================
# PRE-RACE FEATURES ONLY (NO LEAKAGE)
# ============================================================================

logger.info("\n" + "="*80)
logger.info("FEATURE SELECTION: PRE-RACE INFORMATION ONLY")
logger.info("="*80)

# ‚úÖ SAFE: Pre-race features that would be available before a race
pre_race_features = [
    # Raw pre-race metrics
    'qualifying_position',           # ‚úÖ Available before race
    'qualifying_lap_time_s',         # ‚úÖ Available before race
    'elo_before',                    # ‚úÖ Rating before this race
    'elo_final',                     # ‚úÖ Final rating from training
    'overtaking_difficulty',         # ‚úÖ Circuit property (known beforehand)
    'overtaking_difficulty_norm',    # ‚úÖ Normalized version
    
    # Engineered pre-race features (historical, no target leakage)
    'recent_form_5',                 # ‚úÖ Last 5 races (historical)
    'recent_form_3',                 # ‚úÖ Last 3 races (historical)
    'team_perf_trend',               # ‚úÖ Team history (historical)
    'dnf_probability',               # ‚úÖ Historical DNF rate
    'quali_form',                    # ‚úÖ Recent qualifying performance
    'circuit_history_avg_pos',       # ‚úÖ Driver average at this circuit
    'driver_experience',             # ‚úÖ Career races
    'season_progress',               # ‚úÖ Race number in season
    
    # Categorical encodings
    'driver_encoded',
    'team_encoded',
    'circuit_encoded'
]

# ‚ùå REMOVED: Features that leak post-race information
removed_features = {
    'race_lap_time_s': 'Only known after race',
    'laps_completed': 'Only known after race',
    'race_pace_advantage': 'Calculated from race_lap_time_s',
    'lap_completion_pct': 'Calculated from laps_completed',
    'grid_position_advantage': 'Uses finishing_position (target)',
    'position_changes': 'Directly calculated from target',
    'points': 'Only known after race',
    'status': 'Only known after race',
    'elo_delta': 'Change during/after race'
}

logger.info(f"\n‚úÖ Pre-race features ({len(pre_race_features)}):")
for f in pre_race_features:
    logger.info(f"   {f}")

logger.info(f"\n‚ùå Removed post-race features ({len(removed_features)}):")
for f, reason in removed_features.items():
    logger.info(f"   {f:<30} - {reason}")

# ============================================================================
# PREPARE DATA
# ============================================================================

logger.info("\n" + "="*80)
logger.info("DATA PREPARATION")
logger.info("="*80)

df_model = df.copy()

# Encode categoricals
encoders = {}
for col in ['driver', 'team', 'circuit']:
    le = LabelEncoder()
    df_model[f'{col}_encoded'] = le.fit_transform(df_model[col].astype(str))
    encoders[col] = le

# Prepare features
X = df_model[pre_race_features].copy()
X = X.fillna(X.median(numeric_only=True))

# Prepare targets
y_fin = df_model['finishing_position'].copy()
y_dnf = df_model['dnf'].astype(int).copy()
y_pos = df_model['position_changes'].copy()

logger.info(f"\n‚úì Feature matrix shape: {X.shape}")
logger.info(f"‚úì All NaN filled: {X.isnull().sum().sum() == 0}")

# ============================================================================
# TIME-BASED SPLIT (NO LEAKAGE)
# ============================================================================

logger.info("\n" + "="*80)
logger.info("TRAIN-TEST SPLIT (TEMPORAL, NO LEAKAGE)")
logger.info("="*80)

train_mask = df_model['race_year'] < 2024
test_mask = df_model['race_year'] == 2024

X_train = X[train_mask].reset_index(drop=True)
X_test = X[test_mask].reset_index(drop=True)

y_train_fin = y_fin[train_mask].reset_index(drop=True)
y_test_fin = y_fin[test_mask].reset_index(drop=True)

y_train_dnf = y_dnf[train_mask].reset_index(drop=True)
y_test_dnf = y_dnf[test_mask].reset_index(drop=True)

y_train_pos = y_pos[train_mask].reset_index(drop=True)
y_test_pos = y_pos[test_mask].reset_index(drop=True)

logger.info(f"\nTrain: {X_train.shape[0]} races (2018-2023)")
logger.info(f"Test:  {X_test.shape[0]} races (2024 only)")

# ============================================================================
# SCALE FEATURES
# ============================================================================

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

logger.info("‚úì Features scaled")

# ============================================================================
# MODEL 1: FINISHING POSITION
# ============================================================================

logger.info("\n" + "="*80)
logger.info("MODEL 1: FINISHING POSITION (Regression)")
logger.info("="*80)

logger.info("üöÄ Training...")

model_fin = xgb.XGBRegressor(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.7,
    colsample_bytree=0.7,
    objective='reg:squarederror',
    random_state=42,
    verbosity=0
)

model_fin.fit(X_train_scaled, y_train_fin, verbose=False)

y_pred_fin_train = np.clip(model_fin.predict(X_train_scaled), 1, 20)
y_pred_fin_test = np.clip(model_fin.predict(X_test_scaled), 1, 20)

train_mae = mean_absolute_error(y_train_fin, y_pred_fin_train)
test_mae = mean_absolute_error(y_test_fin, y_pred_fin_test)
test_r2 = r2_score(y_test_fin, y_pred_fin_test)

logger.info(f"\n‚úì Train MAE: {train_mae:.3f} positions")
logger.info(f"‚úì Test MAE:  {test_mae:.3f} positions")
logger.info(f"‚úì Test R¬≤:   {test_r2:.4f}")
logger.info(f"\nInterpretation:")
logger.info(f"  On average, predictions are {test_mae:.1f} finishing positions away")
logger.info(f"  Model explains {test_r2*100:.1f}% of position variance")

# ============================================================================
# MODEL 2: DNF CLASSIFICATION
# ============================================================================

logger.info("\n" + "="*80)
logger.info("MODEL 2: DNF PROBABILITY (Classification)")
logger.info("="*80)

logger.info("üöÄ Training...")

model_dnf = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.7,
    colsample_bytree=0.7,
    random_state=42,
    verbosity=0
)

model_dnf.fit(X_train_scaled, y_train_dnf, verbose=False)

y_pred_dnf_train = model_dnf.predict(X_train_scaled)
y_pred_dnf_test = model_dnf.predict(X_test_scaled)

train_acc = accuracy_score(y_train_dnf, y_pred_dnf_train)
test_acc = accuracy_score(y_test_dnf, y_pred_dnf_test)
test_f1 = f1_score(y_test_dnf, y_pred_dnf_test, zero_division=0)

logger.info(f"\n‚úì Train Accuracy: {train_acc:.1%}")
logger.info(f"‚úì Test Accuracy:  {test_acc:.1%}")
logger.info(f"‚úì Test F1-Score:  {test_f1:.4f}")
logger.info(f"\nInterpretation:")
logger.info(f"  Model correctly predicts finish/DNF {test_acc*100:.1f}% of races")

# ============================================================================
# MODEL 3: POSITION CHANGES
# ============================================================================

logger.info("\n" + "="*80)
logger.info("MODEL 3: POSITION CHANGES (Regression)")
logger.info("="*80)

logger.info("üöÄ Training...")

model_pos = xgb.XGBRegressor(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.7,
    colsample_bytree=0.7,
    objective='reg:squarederror',
    random_state=42,
    verbosity=0
)

model_pos.fit(X_train_scaled, y_train_pos, verbose=False)

y_pred_pos_train = model_pos.predict(X_train_scaled)
y_pred_pos_test = model_pos.predict(X_test_scaled)

train_mae_pos = mean_absolute_error(y_train_pos, y_pred_pos_train)
test_mae_pos = mean_absolute_error(y_test_pos, y_pred_pos_test)
test_r2_pos = r2_score(y_test_pos, y_pred_pos_test)

logger.info(f"\n‚úì Train MAE: {train_mae_pos:.3f} positions")
logger.info(f"‚úì Test MAE:  {test_mae_pos:.3f} positions")
logger.info(f"‚úì Test R¬≤:   {test_r2_pos:.4f}")

# ============================================================================
# FEATURE IMPORTANCE
# ============================================================================

logger.info("\n" + "="*80)
logger.info("FEATURE IMPORTANCE (Top 10)")
logger.info("="*80)

importance_dict = model_fin.get_booster().get_score(importance_type='weight')
importance_df = pd.DataFrame([
    {'feature': pre_race_features[i], 'importance': importance_dict.get(f'f{i}', 0)}
    for i in range(len(pre_race_features))
]).sort_values('importance', ascending=False)

for idx, row in importance_df.head(10).iterrows():
    logger.info(f"  {row['feature']:30s} : {row['importance']:.0f}")

# ============================================================================
# SAVE MODELS
# ============================================================================

import os
import pickle

os.makedirs(MODEL_OUTPUT, exist_ok=True)

model_fin.save_model(f'{MODEL_OUTPUT}/finishing_position_model.json')
model_dnf.save_model(f'{MODEL_OUTPUT}/dnf_model.json')
model_pos.save_model(f'{MODEL_OUTPUT}/position_changes_model.json')

with open(f'{MODEL_OUTPUT}/scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open(f'{MODEL_OUTPUT}/encoders.pkl', 'wb') as f:
    pickle.dump(encoders, f)

with open(f'{MODEL_OUTPUT}/feature_names.pkl', 'wb') as f:
    pickle.dump(pre_race_features, f)

logger.info(f"\nüíæ Models saved to {MODEL_OUTPUT}/")

# ============================================================================
# SUMMARY
# ============================================================================

logger.info("\n" + "="*80)
logger.info("‚úÖ MODEL TRAINING COMPLETE (NO DATA LEAKAGE)")
logger.info("="*80)

summary = f"""
REALISTIC MODEL PERFORMANCE (No Data Leakage):

Finishing Position:
  Test MAE: {test_mae:.3f} positions (realistic for F1 predictions)
  Test R¬≤:  {test_r2:.4f}

DNF Prediction:
  Test Accuracy: {test_acc:.1%}
  Test F1-Score: {test_f1:.4f}

Position Changes:
  Test MAE: {test_mae_pos:.3f} positions
  Test R¬≤:  {test_r2_pos:.4f}

Features: {len(pre_race_features)} (all pre-race, no target leakage)
Train: {len(X_train)} races (2018-2023)
Test:  {len(X_test)} races (2024)
"""

logger.info(summary)

print("\n‚úÖ These scores are REALISTIC, not miracles!")
print(f"   Test MAE {test_mae:.2f} positions means predictions are ¬±{test_mae:.1f} place off")
print(f"   Test F1 {test_f1:.2f} means {test_f1*100:.0f}% balanced DNF prediction")
````

````python
import fastf1
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import logging
from datetime import datetime


# ============================================================================
# LOGGING SETUP
# ============================================================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)


# =================================================================================================================================
# CONFIG
# ============================================================================
CACHE_DIR = r'C:\Users\moina\f1_cache'
OUTPUT_FILE = 'f1_training_dataset_2018_2024.csv'
YEARS = [2018, 2019, 2020, 2021, 2022, 2023, 2024]


# Enable FastF1 cache
fastf1.Cache.enable_cache(CACHE_DIR)


# ============================================================================
# UTILITY FUNCTIONS (FINAL CORRECTED VERSION - COLUMN NAMES FIXED)
# ============================================================================


def get_col(df, possible_names, default=None):
    """
    Safely get column name with fallback names.
    Handles FastF1 column name variations across versions.
    
    Args:
        df: DataFrame or results object with .columns attribute
        possible_names: List of possible column names to try
        default: Default value if no column found
    
    Returns:
        Column name if found, else default
    """
    if not hasattr(df, 'columns'):
        return default
    
    for name in possible_names:
        if name in df.columns:
            return name
    
    return default


def get_session_safe(year, event_name):
    """
    Try to get qualifying and race sessions independently.
    FastF1 v3.6.1 column structure requires Abbreviation (not Driver in results).
    
    Args:
        year: F1 season
        event_name: Event name from schedule
    
    Returns:
        tuple: (qualifying_session or None, race_session or None)
    """
    q_types = ['Q', 'SQ', 'S']  # Try Qualifying ‚Üí Sprint Q ‚Üí Sprint
    r_types = ['R']
    
    q_session = None
    r_session = None
    
    # Try qualifying independently
    for q_type in q_types:
        try:
            q_session = fastf1.get_session(year, event_name, q_type)
            logger.debug(f"‚úì Got {q_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {q_type} failed: {type(e).__name__}")
            continue
    
    if q_session is None:
        logger.debug(f"No qualifying session found for {event_name}")
    
    # Try race independently (regardless of qual)
    for r_type in r_types:
        try:
            r_session = fastf1.get_session(year, event_name, r_type)
            logger.debug(f"‚úì Got {r_type} session for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {r_type} failed: {type(e).__name__}")
            continue
    
    if r_session is None:
        logger.debug(f"No race session found for {event_name}")
    
    return q_session, r_session


def classify_dnf_corrected(status):
    """
    CORRECTED DNF Classification (verified against 2018 F1 data).
    
    Returns 1 if driver Did Not Finish (not classified in final results)
    Returns 0 if driver completed race and was classified (regardless of laps down)
    Returns np.nan if status is unknown/ambiguous
    
    Logic:
    - 'Finished' ‚Üí DNF=0 (classified, completed lead lap)
    - '+1 Lap', '+2 Laps', etc. ‚Üí DNF=0 (classified, lapped but finished)
    - 'Engine', 'Power loss', 'Collision', etc. ‚Üí DNF=1 (retired, not classified)
    - Unknown/ambiguous ‚Üí NaN (don't assume)
    
    Args:
        status: Status value from FastF1 results
    
    Returns:
        0 if finished/classified, 1 if DNF/not classified, NaN if unknown
    """
    if pd.isna(status):
        return np.nan  # Changed from 0 - unknown should be NaN
    
    status_str = str(status).strip().lower()
    
    # CLASSIFIED DRIVERS (DNF = 0)
    # 1. Drivers who finished on lead lap
    if status_str == 'finished':
        return 0
    
    # 2. Drivers who finished but were lapped (still classified)
    if '+' in status_str and 'lap' in status_str:
        # Matches: '+1 lap', '+2 laps', '+3 lap', etc.
        # KEY FIX: '+1 Lap' is CLASSIFIED, not DNF
        return 0
    
    # NOT CLASSIFIED DRIVERS (DNF = 1)
    dnf_keywords = ['engine', 'collision', 'accident', 'gearbox', 'hydraulics', 
                    'retired', 'spun off', 'damage', 'electronics', 'brakes',
                    'power loss', 'fuel system', 'transmission', 'suspension']
    if any(kw in status_str for kw in dnf_keywords):
        return 1
    
    # Unknown/ambiguous - return NaN instead of assuming finished
    return np.nan


def get_event_date_safe(event_row, schedule_idx):
    """
    Extract event date with multiple fallbacks.
    
    Args:
        event_row: Row from FastF1 schedule
        schedule_idx: Position in schedule
    
    Returns:
        datetime or NaT
    """
    date_columns = ['EventDate', 'Session5Date', 'RaceDate', 'Date']
    
    for col in date_columns:
        if col in event_row.index:
            date_val = event_row.get(col, pd.NaT)
            if pd.notna(date_val):
                return pd.to_datetime(date_val, errors='coerce')
    
    logger.debug(f"No valid date found in schedule row {schedule_idx}")
    return pd.NaT


def get_team_perf_score(df_race_results):
    """
    Calculate team performance from actual race results.
    
    Args:
        df_race_results: DataFrame with 'team' and 'finishing_position' columns
    
    Returns:
        dict: team -> performance_score (0-1)
    """
    team_finishes = df_race_results[df_race_results['finishing_position'].notna()].groupby('team').size()
    if len(team_finishes) == 0:
        return {}
    max_finishes = team_finishes.max()
    team_perf = (team_finishes / max_finishes).to_dict()
    return team_perf


def calculate_elo_safe(results_df, k=24, initial_elo=1500):
    """
    Calculate Elo ratings with robust schedule_idx handling.
    
    FIXED: Uses 'finishing_position' column name (matches actual data)
    
    Args:
        results_df: DataFrame with race results
        k: Elo K-factor (default 24)
        initial_elo: Starting Elo rating (default 1500)
    
    Returns:
        tuple: (elo_history_df, final_elo_df)
    """
    res = results_df.copy()
    
    # FIXED: Use correct column name 'finishing_position'
    res["finishing_position"] = pd.to_numeric(res["finishing_position"], errors="coerce")
    res = res[res["finishing_position"].notna()].copy()

    if len(res) == 0:
        logger.warning("No finished races found for Elo calculation")
        return pd.DataFrame(), pd.DataFrame()

    # Generate per-year ordering if schedule_idx missing
    if 'schedule_idx' not in res.columns or res['schedule_idx'].isna().all():
        logger.warning("schedule_idx missing ‚Äî generating per-year ordering")
        res['schedule_idx'] = res.groupby('race_year').cumcount()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    else:
        missing_idx = res['schedule_idx'].isna().sum()
        if missing_idx > 0:
            logger.warning(f"schedule_idx has {missing_idx} NaN values ‚Äî filling with per-year rank")
            res.loc[res['schedule_idx'].isna(), 'schedule_idx'] = \
                res[res['schedule_idx'].isna()].groupby('race_year').cumcount() + res['schedule_idx'].max()
        sort_keys = ["race_year", "event_date", "schedule_idx"]
    
    # Build race_order
    race_order = res[["race_year", "race_key", "event_date", "schedule_idx"]].drop_duplicates()
    race_order = race_order.sort_values(sort_keys).reset_index(drop=True)
    
    order_map = {rk: i for i, rk in enumerate(race_order["race_key"].tolist())}
    res["race_order_idx"] = res["race_key"].map(order_map)
    
    # Fail loudly if mapping failed
    unmapped = res[res["race_order_idx"].isna()]["race_key"].unique()
    if len(unmapped) > 0:
        raise ValueError(f"‚ùå CRITICAL: These race_keys were not mapped to order: {unmapped}")
    
    # FIXED: Use correct column name 'finishing_position'
    res = res.sort_values(["race_order_idx", "finishing_position"]).reset_index(drop=True)

    elo = {}
    history = []

    for rk in res["race_key"].unique():
        race_rows = res[res["race_key"] == rk].sort_values("finishing_position")
        drivers_in_race = race_rows["driver"].tolist()
        positions = race_rows["finishing_position"].astype(int).tolist()
        n = len(drivers_in_race)

        if n <= 0:
            continue

        for d in drivers_in_race:
            elo.setdefault(d, initial_elo)

        deltas = {}
        for i, d_i in enumerate(drivers_in_race):
            elo_i = elo[d_i]
            exp_sum = 0.0
            for j, d_j in enumerate(drivers_in_race):
                if d_i == d_j:
                    continue
                diff = elo[d_j] - elo_i
                p = 1.0 / (1.0 + 10 ** (diff / 400.0))
                exp_sum += p
            expected_norm = exp_sum / (n - 1) if n > 1 else 0.0
            actual_norm = (n - positions[i]) / (n - 1) if n > 1 else 0.0
            delta = k * (actual_norm - expected_norm)
            deltas[d_i] = delta

        for d in drivers_in_race:
            history.append({
                "race_key": rk,
                "driver": d,
                "elo_before": elo[d],
                "elo_delta": deltas.get(d, 0)
            })

        for d, delta in deltas.items():
            elo[d] = elo[d] + delta

    final_elo = pd.DataFrame([(d, e) for d, e in elo.items()], columns=['driver', 'elo_final'])
    return pd.DataFrame(history), final_elo


def calculate_overtaking_difficulty_from_data(df):
    """
    Calculate overtaking difficulty based on average position changes per circuit
    
    FIXED: Uses 'position_changes' column name (matches actual data)
    """
    # FIXED: Use correct column name 'position_changes'
    if 'position_changes' not in df.columns:
        logger.warning("position_changes column not found.")
        return {}
    
    circuit_overtaking = df.groupby('circuit').agg({
        'position_changes': 'mean'
    }).dropna()
    
    if len(circuit_overtaking) == 0:
        logger.warning("No valid position_changes data for circuits.")
        return {}
    
    max_changes = circuit_overtaking['position_changes'].max()
    if max_changes > 0:
        circuit_overtaking['difficulty'] = 1 - (circuit_overtaking['position_changes'] / max_changes)
    else:
        circuit_overtaking['difficulty'] = 0.6
    
    return circuit_overtaking['difficulty'].to_dict()


# ============================================================================
# LOAD DATA FROM LOCAL CACHE
# ============================================================================

logger.info("="*80)
logger.info("LOADING F1 DATA FROM LOCAL CACHE - 2018-2024")
logger.info("="*80)
logger.info(f"Cache directory: {CACHE_DIR}\n")

all_rows = []
successful_years = []
failed_years = []

for year in YEARS:
    logger.info(f"\nüìÖ Processing season {year}")
    logger.info("-"*80)
    
    try:
        schedule = fastf1.get_event_schedule(year)
        logger.info(f"Found {len(schedule)} events for {year}")
    except Exception as e:
        logger.error(f"Failed to load schedule: {e}")
        failed_years.append((year, str(e)))
        continue
    
    year_rows = 0
    
    for schedule_idx, (_, event_row) in enumerate(schedule.iterrows()):
        try:
            event_name = event_row.get('EventName') or event_row.get('Event') or 'Unknown'
            circuit_name = event_row.get('OfficialEventName') or event_name
            event_date = get_event_date_safe(event_row, schedule_idx)
            
            race_key = f"{year}__{schedule_idx:02d}__{event_name}".replace(" ", "_").replace("/", "_")
            
            # Get sessions independently
            q, r = get_session_safe(year, event_name)
            
            # Skip only if BOTH missing
            if q is None and r is None:
                logger.debug(f"Skipping {event_name}: no sessions found")
                continue
            
            # Load each session that exists
            if q is not None:
                try:
                    q.load(laps=True, telemetry=False, weather=False)
                except Exception as q_load_err:
                    logger.debug(f"Failed to load qualifying for {event_name}: {q_load_err}")
                    q = None
            
            if r is not None:
                try:
                    r.load(laps=True, telemetry=False, weather=False)
                except Exception as r_load_err:
                    logger.debug(f"Failed to load race for {event_name}: {r_load_err}")
                    r = None
            
            # Skip if BOTH ended up None after load attempts
            if q is None and r is None:
                logger.debug(f"Skipping {event_name}: failed to load either session")
                continue
            
            # Extract qualifying data (using defensive column names)
            q_data = {}
            q_lap_times = {}
            
            if q is not None:
                if q.results is not None and not q.results.empty:
                    try:
                        # Get driver column name with fallback
                        driver_col = get_col(q.results, ['Abbreviation', 'Driver', 'DriverNumber'])
                        team_col = get_col(q.results, ['TeamName', 'Team', 'Constructor'])
                        pos_col = get_col(q.results, ['Position', 'GridPosition', 'P'])
                        status_col = get_col(q.results, ['Status', 'Retired'])
                        
                        if driver_col:
                            for idx, row in q.results.iterrows():
                                driver = row[driver_col]
                                q_data[driver] = {
                                    'team': row.get(team_col, np.nan) if team_col else np.nan,
                                    'qualifying_position': row.get(pos_col, np.nan) if pos_col else np.nan,
                                    'q_status': row.get(status_col, np.nan) if status_col else np.nan
                                }
                        else:
                            logger.debug(f"Failed to find driver column in qualifying results for {event_name}")
                    except Exception as q_extract_err:
                        logger.debug(f"Failed to extract qualifying data for {event_name}: {q_extract_err}")
                
                # Extract qualifying lap times with type guard
                try:
                    if not q.laps.empty:
                        driver_col_laps = get_col(q.laps, ['Driver', 'Abbreviation', 'DriverNumber'])
                        laptime_col = get_col(q.laps, ['LapTime', 'Time'])
                        if driver_col_laps and laptime_col:
                            lap_times_td = q.laps.groupby(driver_col_laps)[laptime_col].mean()
                            # CRITICAL: Check if result is timedelta before calling .dt
                            # Prevents AttributeError if all laps are NaT or column contains strings
                            if pd.api.types.is_timedelta64_dtype(lap_times_td):
                                q_lap_times = lap_times_td.dt.total_seconds().to_dict()
                            else:
                                logger.debug(f"Qualifying lap times not timedelta type for {event_name} (likely red flag session)")
                                q_lap_times = {}
                except Exception as q_laps_err:
                    logger.debug(f"Failed to extract qualifying lap times: {q_laps_err}")
                    q_lap_times = {}
            
            # Extract race data (using defensive column names)
            r_data = {}
            r_lap_times = {}
            
            if r is not None:
                if r.results is not None and not r.results.empty:
                    try:
                        # Get driver column name with fallback
                        driver_col = get_col(r.results, ['Abbreviation', 'Driver', 'DriverNumber'])
                        pos_col = get_col(r.results, ['Position', 'GridPosition', 'P'])
                        status_col = get_col(r.results, ['Status', 'Retired'])
                        points_col = get_col(r.results, ['Points', 'Pts', 'Points'])
                        laps_col = get_col(r.results, ['Laps', 'Lap', 'TotalLaps'])
                        
                        if driver_col:
                            for idx, row in r.results.iterrows():
                                driver = row[driver_col]
                                r_data[driver] = {
                                    'finishing_position': row.get(pos_col, np.nan) if pos_col else np.nan,
                                    'r_status': row.get(status_col, np.nan) if status_col else np.nan,
                                    'points': row.get(points_col, np.nan) if points_col else np.nan,
                                    'laps_completed': row.get(laps_col, np.nan) if laps_col else np.nan
                                }
                        else:
                            logger.debug(f"Failed to find driver column in race results for {event_name}")
                    except Exception as r_extract_err:
                        logger.debug(f"Failed to extract race data for {event_name}: {r_extract_err}")
                
                # Extract race lap times with type guard
                try:
                    if not r.laps.empty:
                        driver_col_laps = get_col(r.laps, ['Driver', 'Abbreviation', 'DriverNumber'])
                        laptime_col = get_col(r.laps, ['LapTime', 'Time'])
                        if driver_col_laps and laptime_col:
                            lap_times_td = r.laps.groupby(driver_col_laps)[laptime_col].mean()
                            # CRITICAL: Check if result is timedelta before calling .dt
                            # Prevents AttributeError if all laps are NaT or column contains strings
                            if pd.api.types.is_timedelta64_dtype(lap_times_td):
                                r_lap_times = lap_times_td.dt.total_seconds().to_dict()
                            else:
                                logger.debug(f"Race lap times not timedelta type for {event_name} (likely red flag session)")
                                r_lap_times = {}
                except Exception as r_laps_err:
                    logger.debug(f"Failed to extract race lap times: {r_laps_err}")
                    r_lap_times = {}
            
            # Skip if no data
            if not q_data and not r_data:
                logger.debug(f"No data extracted for {event_name}")
                continue
            
            # Combine all drivers
            all_drivers = set(q_data.keys()).union(r_data.keys())
            
            for driver in all_drivers:
                q_info = q_data.get(driver, {})
                r_info = r_data.get(driver, {})
                
                qual_pos = q_info.get('qualifying_position', np.nan)
                fin_pos = r_info.get('finishing_position', np.nan)
                r_status = r_info.get('r_status', np.nan)
                
                # CORRECTED DNF classification
                dnf_flag = classify_dnf_corrected(r_status)
                dnf_known = pd.notna(dnf_flag)
                
                # Calculate position changes
                if pd.notna(qual_pos) and pd.notna(fin_pos):
                    position_changes = abs(fin_pos - qual_pos)
                else:
                    position_changes = np.nan
                
                row = {
                    "race_key": race_key,
                    "race_year": year,
                    "schedule_idx": schedule_idx,
                    "event": event_name,
                    "circuit": circuit_name,
                    "event_date": event_date,
                    "driver": driver,  # Abbreviation
                    "team": q_info.get('team', np.nan),
                    "qualifying_position": qual_pos,
                    "qualifying_lap_time_s": q_lap_times.get(driver, np.nan),
                    "race_lap_time_s": r_lap_times.get(driver, np.nan),
                    "finishing_position": fin_pos,
                    "position_changes": position_changes,
                    "dnf": dnf_flag,
                    "dnf_known": dnf_known,
                    "status": r_status,
                    "points": r_info.get('points', np.nan),
                    "laps_completed": r_info.get('laps_completed', np.nan)
                }
                all_rows.append(row)
                year_rows += 1
        
        except Exception as e:
            logger.error(f"Unexpected error processing {event_row.get('EventName', 'Unknown')}: {e}")
            continue
    
    if year_rows > 0:
        successful_years.append((year, year_rows))
        logger.info(f"‚úì Loaded {year_rows} rows from {year}")
    else:
        logger.info(f"‚úó Year {year} produced no rows")

logger.info(f"\n‚úÖ Extraction complete:")
logger.info(f"  Successful years: {successful_years}")
logger.info(f"  Total rows collected: {len(all_rows)}")

# ============================================================================
# CONVERT TO DATAFRAME & ENFORCE SCHEMA
# ============================================================================

if len(all_rows) == 0:
    logger.error("‚ùå No data loaded! Cannot proceed.")
else:
    logger.info(f"\nüìä Converting {len(all_rows)} rows to DataFrame...")
    df_raw = pd.DataFrame(all_rows)
    
    logger.info(f"\n{'='*80}")
    logger.info("SCHEMA VALIDATION & DTYPE ENFORCEMENT")
    logger.info(f"{'='*80}")
    
    # Define expected schema
    schema = {
        'race_year': 'int64',
        'schedule_idx': 'Int64',  # Nullable int
        'qualifying_position': 'float64',
        'finishing_position': 'float64',
        'qualifying_lap_time_s': 'float64',
        'race_lap_time_s': 'float64',
        'position_changes': 'float64',
        'dnf': 'float64',  # Allow NaN now
        'dnf_known': 'bool',
        'points': 'float64',
        'laps_completed': 'float64'
    }
    
    # Enforce dtypes
    logger.info("\nüîç Enforcing column dtypes...")
    for col, dtype in schema.items():
        if col in df_raw.columns:
            try:
                df_raw[col] = df_raw[col].astype(dtype)
                logger.debug(f"  ‚úì {col:30s} ‚Üí {dtype}")
            except Exception as e:
                logger.error(f"  ‚úó Failed to convert {col} to {dtype}: {e}")
                logger.error(f"    Sample values: {df_raw[col].head().tolist()}")
        else:
            logger.warning(f"  ‚ö† Column {col} not found in DataFrame")
    
    # Validate critical columns
    logger.info("\n‚úì Validating critical columns...")
    try:
        assert df_raw['race_year'].notna().all(), "Missing race_year values"
        logger.info("  ‚úì race_year: no missing values")
        
        assert df_raw['driver'].notna().all(), "Missing driver values"
        logger.info("  ‚úì driver: no missing values")
        
        assert df_raw['dnf'].isna().sum() + df_raw['dnf'].isin([0, 1, np.nan]).sum() == len(df_raw), "Invalid dnf values (must be 0, 1, or NaN)"
        logger.info("  ‚úì dnf: all values are 0, 1, or NaN")
        
        logger.info(f"\n‚úÖ Schema validation PASSED")
    except AssertionError as e:
        logger.error(f"\n‚ùå Schema validation FAILED: {e}")
        raise
    
    logger.info(f"\n{'='*80}")
    logger.info("üìä DATASET SUMMARY AFTER VALIDATION")
    logger.info(f"{'='*80}")
    logger.info(f"Shape: {df_raw.shape}")
    logger.info(f"Years: {sorted(df_raw['race_year'].unique())}")
    logger.info(f"Drivers: {df_raw['driver'].nunique()}")
    logger.info(f"Circuits: {df_raw['circuit'].nunique()}")
    logger.info(f"DNF=0 (Finished): {(df_raw['dnf'] == 0).sum()}")
    logger.info(f"DNF=1 (Not Classified): {(df_raw['dnf'] == 1).sum()}")
    logger.info(f"DNF=NaN (Unknown): {df_raw['dnf'].isna().sum()}")
    logger.info(f"{'='*80}\n")
    
    finish_rate = (df_raw['dnf'] == 0).sum() / len(df_raw) * 100 if len(df_raw) > 0 else 0
    print(f"\n‚úÖ Pipeline complete! {len(df_raw)} rows with validated schema")
    print(f"üìä Finish Rate: {finish_rate:.1f}% (should be ~70-80% for typical F1 season)")
````

````python
# =========================
# FEATURE ENGINEERING (fixed)
# =========================

print("üîß Feature Engineering...")
print()

df = pd.DataFrame(all_rows)

# --- Normalise column names to expected snake_case used earlier ---
df = df.rename(columns={
    "QualifyingPosition": "qualifying_position",
    "FinishingPosition": "finishing_position",
    "QualifyingMeanLap_s": "qualifying_lap_time_s",
    "AverageRaceLap_s": "race_lap_time_s",
    "DNF": "dnf_flag",
    "Status": "status",
    "PositionChanges": "position_changes",
    "Team": "team",
    "Event": "event",
    "Circuit": "circuit",
    "Driver": "driver"
}, errors="ignore")

# --- Ensure core columns exist (avoid KeyError later) ---
expected_cols = [
    "race_key", "race_year", "event_date", "race_order_idx",
    "event", "circuit", "driver", "team",
    "qualifying_position", "finishing_position",
    "qualifying_lap_time_s", "race_lap_time_s",
    "position_changes", "dnf", "status", "points", "laps_completed"
]
for c in expected_cols:
    if c not in df.columns:
        df[c] = np.nan

# --- Coerce numeric columns ---
num_cols = ["qualifying_position", "finishing_position",
            "qualifying_lap_time_s", "race_lap_time_s",
            "position_changes", "points", "laps_completed"]
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# --- Ensure event_date is datetime ---
df["event_date"] = pd.to_datetime(df.get("event_date", pd.NaT), errors="coerce")

# --- Build robust race ordering if not present ---
if "race_order_idx" not in df.columns or df["race_order_idx"].isna().all():
    df = df.sort_values(["race_year", "event_date", "race_key"]).reset_index(drop=True)
    # generate a per-year incremental order
    tmp = df[["race_year", "event_date", "race_key"]].drop_duplicates()
    tmp = tmp.sort_values(["race_year", "event_date"]).reset_index(drop=True)
    tmp["race_order_idx"] = tmp.groupby("race_year").cumcount()
    order_map = dict(zip(tmp["race_key"], tmp["race_order_idx"]))
    df["race_order_idx"] = df["race_key"].map(order_map)

# --- Sort for all time-based transforms: chronological per driver ---
df = df.sort_values(["driver", "race_year", "race_order_idx", "event_date"]).reset_index(drop=True)

# -----------------------
# 1) Elo Rating
# -----------------------
print("  1. Calculating Elo Ratings...")
# ensure you're calling the correct Elo function (history, final)
elo_history, elo_final = calculate_elo_safe(df, k=24, initial_elo=1500)  # use safe version
# elo_history should have columns: race_key, driver, elo_before
if not elo_history.empty and {"race_key","driver","elo_before"}.issubset(elo_history.columns):
    df = df.merge(elo_history[["race_key", "driver", "elo_before"]],
                  on=["race_key", "driver"], how="left")
    df = df.rename(columns={"elo_before": "EloRating"})
    df["EloRating"] = df["EloRating"].fillna(1500)
else:
    df["EloRating"] = 1500

# -----------------------
# 2) Team Performance Score (per-year)
# -----------------------
print("  2. Calculating Team Performance Scores...")
team_perf_all = {}
for year in df['race_year'].dropna().unique():
    year_data = df[df['race_year'] == year]
    # get_team_perf_score expects a single df argument and returns dict
    team_perf_all.update(get_team_perf_score(year_data))
df["TeamPerfScore"] = df["team"].map(team_perf_all).fillna(0.5)

# -----------------------
# 3) Recent Form (last 5 races) - NO DATA LEAKAGE
# -----------------------
print("  3. Calculating Recent Form (last 5 races)...")
# operate on numeric finishing_position
df["RecentFormAvg"] = (
    df.groupby("driver")["finishing_position"]
      .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)
# fallback fill: use median of finishing_position (numeric)
median_fin = df["finishing_position"].median(skipna=True)
df["RecentFormAvg"] = df["RecentFormAvg"].fillna(median_fin)

# -----------------------
# 4) Circuit History - NO DATA LEAKAGE
# -----------------------
print("  4. Calculating Circuit-specific History...")
df["CircuitHistoryAvg"] = (
    df.groupby(["driver", "circuit"])["finishing_position"]
      .transform(lambda x: x.shift(1).expanding(min_periods=1).mean())
)
df["CircuitHistoryAvg"] = df["CircuitHistoryAvg"].fillna(median_fin)

# -----------------------
# 5) Total Races (driver experience) - NO DATA LEAKAGE
# -----------------------
print("  5. Calculating Driver Experience (total races)...")
# cumcount now represents how many prior races the driver had
df["TotalRaces"] = df.groupby("driver").cumcount()

# -----------------------
# 6) Driver Experience Score
# -----------------------
print("  6. Calculating Driver Experience Score...")
max_r = df["TotalRaces"].max()
df["DriverExperienceScore"] = (df["TotalRaces"] / max_r).fillna(0) if max_r and max_r > 0 else 0.0

# -----------------------
# 7) Overtaking Difficulty (position_changes must match expected column)
# -----------------------
print("  7. Calculating Dynamic Overtaking Difficulty...")
# ensure position_changes exists and numeric
if "position_changes" not in df.columns or df["position_changes"].isna().all():
    # compute from qual/finish only where both available
    df["position_changes"] = (df["qualifying_position"] - df["finishing_position"]).abs()
overtaking_dict = calculate_overtaking_difficulty_from_data(df)  # expects 'position_changes' lower-case
df["OvertakingDifficulty"] = df["circuit"].map(overtaking_dict).fillna(0.6)

# -----------------------
# 8) Clean Air Pace (from qualifying vs race avg)
# -----------------------
print("  8. Calculating Clean Air Pace Metric...")
# use per-row lap times; guard divide by zero / NaN
df["CleanAirPace"] = df["qualifying_lap_time_s"] / df["race_lap_time_s"]
df["CleanAirPace"] = df["CleanAirPace"].replace([np.inf, -np.inf], np.nan).fillna(1.0)
df["CleanAirPace"] = df["CleanAirPace"].clip(lower=0.5, upper=1.5)  # wider clip to avoid hard truncation

# -----------------------
# 9) DNF Probability
# -----------------------
print("  9. Extracting DNF information...")
# keep original dnf flag column if present; create numeric probability (0/1/NaN)
if "dnf" in df.columns:
    df["DNFProbability"] = pd.to_numeric(df["dnf"], errors="coerce")
elif "dnf_flag" in df.columns:
    df["DNFProbability"] = pd.to_numeric(df["dnf_flag"], errors="coerce")
else:
    df["DNFProbability"] = np.nan

# -----------------------
# 10) Position Changes (already)
# -----------------------
print(" 10. Position Changes (already calculated)...")
# ensure column exists and numeric
df["PositionChanges"] = pd.to_numeric(df.get("position_changes", np.nan), errors="coerce")

# -----------------------
# Final column mapping - ensure final columns exist in df
# -----------------------
final_columns = [
    "race_key", "race_year", "event", "circuit", "driver", "team",
    "qualifying_position", "EloRating", "TeamPerfScore",
    "RecentFormAvg", "CircuitHistoryAvg",
    "qualifying_lap_time_s", "race_lap_time_s", "CleanAirPace",
    "DriverExperienceScore", "OvertakingDifficulty", "TotalRaces",
    "DNFProbability", "PositionChanges",
    "finishing_position", "dnf", "status"
]

# ensure missing final cols are created to avoid KeyError
for c in final_columns:
    if c not in df.columns:
        df[c] = np.nan

df_final = df[final_columns].copy()

# Save to CSV
df_final.to_csv(OUTPUT_FILE, index=False)

print(f"\n{'='*80}")
print(f"‚úÖ DATASET SAVED: {OUTPUT_FILE}")
print(f"Shape: {df_final.shape}")
print(f"Years: {sorted(df_final['race_year'].dropna().unique())}")
print(f"{'='*80}\n")

print(df_final.head(10))
print()
print(df_final.info())
````

````python
import fastf1
import pandas as pd
import numpy as np
import logging
from datetime import datetime

# ============================================================================
# LOGGING SETUP
# ============================================================================
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# ============================================================================
# CONFIG - TEST WITH JUST 2018
# ============================================================================
CACHE_DIR = r'C:\Users\moina\f1_cache'
OUTPUT_FILE = 'f1_test_2018_only.csv'
YEARS = [2018]  # TEST ONLY 2018

fastf1.Cache.enable_cache(CACHE_DIR)

# ============================================================================
# MINIMAL VERSION OF MAIN FUNCTIONS
# ============================================================================

def get_session_safe(year, event_name):
    """Get Q and R independently"""
    q_types = ['Q', 'SQ', 'S']
    r_types = ['R']
    
    q_session = None
    r_session = None
    
    for q_type in q_types:
        try:
            q_session = fastf1.get_session(year, event_name, q_type)
            logger.debug(f"‚úì Got {q_type} for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {q_type} failed: {type(e).__name__}")
            continue
    
    for r_type in r_types:
        try:
            r_session = fastf1.get_session(year, event_name, r_type)
            logger.debug(f"‚úì Got {r_type} for {event_name}")
            break
        except Exception as e:
            logger.debug(f"  {r_type} failed: {type(e).__name__}")
            continue
    
    return q_session, r_session


def classify_dnf(status):
    """Classify DNF"""
    if pd.isna(status):
        return np.nan
    if str(status).strip().lower() == 'finished':
        return 0
    return 1


# ============================================================================
# LOAD DATA
# ============================================================================

all_rows = []

for year in YEARS:
    logger.info(f"\nüìÖ Processing season {year}")
    
    try:
        schedule = fastf1.get_event_schedule(year)
        logger.info(f"Found {len(schedule)} events")
    except Exception as e:
        logger.error(f"Failed to load schedule: {e}")
        continue
    
    year_rows = 0
    
    for schedule_idx, (_, event_row) in enumerate(schedule.iterrows()):
        event_name = event_row.get('EventName') or 'Unknown'
        
        try:
            q, r = get_session_safe(year, event_name)
            
            if q is None and r is None:
                logger.info(f"  ‚è≠Ô∏è  Skipping {event_name}: no sessions")
                continue
            
            # Load sessions
            if q is not None:
                try:
                    q.load(laps=True, telemetry=False, weather=False)
                except Exception as e:
                    logger.warning(f"  Failed to load Q for {event_name}: {e}")
                    q = None
            
            if r is not None:
                try:
                    r.load(laps=True, telemetry=False, weather=False)
                except Exception as e:
                    logger.warning(f"  Failed to load R for {event_name}: {e}")
                    r = None
            
            if q is None and r is None:
                logger.info(f"  ‚è≠Ô∏è  {event_name}: both sessions failed to load")
                continue
            
            # Extract qualifying
            q_data = {}
            if q is not None and q.results is not None and not q.results.empty:
                try:
                    for idx, row in q.results.iterrows():
                        driver = row['Abbreviation']
                        q_data[driver] = {
                            'team': row.get('TeamName', np.nan),
                            'qual_pos': row.get('Position', np.nan)
                        }
                except Exception as e:
                    logger.warning(f"  Failed to extract Q data for {event_name}: {e}")
            
            # Extract race
            r_data = {}
            if r is not None and r.results is not None and not r.results.empty:
                try:
                    for idx, row in r.results.iterrows():
                        driver = row['Abbreviation']
                        r_data[driver] = {
                            'fin_pos': row.get('Position', np.nan),
                            'status': row.get('Status', np.nan),
                            'points': row.get('Points', np.nan)
                        }
                except Exception as e:
                    logger.warning(f"  Failed to extract R data for {event_name}: {e}")
            
            if not q_data and not r_data:
                logger.info(f"  ‚è≠Ô∏è  {event_name}: no data extracted")
                continue
            
            # Combine
            all_drivers = set(q_data.keys()).union(r_data.keys())
            
            for driver in all_drivers:
                q_info = q_data.get(driver, {})
                r_info = r_data.get(driver, {})
                
                qual_pos = q_info.get('qual_pos', np.nan)
                fin_pos = r_info.get('fin_pos', np.nan)
                status = r_info.get('status', np.nan)
                
                dnf = classify_dnf(status)
                
                if pd.notna(qual_pos) and pd.notna(fin_pos):
                    pos_changes = abs(fin_pos - qual_pos)
                else:
                    pos_changes = np.nan
                
                row = {
                    'year': year,
                    'event': event_name,
                    'driver': driver,
                    'team': q_info.get('team', np.nan),
                    'qual_pos': qual_pos,
                    'fin_pos': fin_pos,
                    'pos_changes': pos_changes,
                    'dnf': dnf,
                    'status': status,
                    'points': r_info.get('points', np.nan)
                }
                all_rows.append(row)
                year_rows += 1
            
            logger.info(f"  ‚úì {event_name}: extracted {len(all_drivers)} drivers")
        
        except Exception as e:
            logger.error(f"  ‚ùå {event_name}: {e}")
            continue
    
    logger.info(f"‚úì Year {year}: {year_rows} total rows")

# ============================================================================
# SAVE & VERIFY
# ============================================================================

logger.info(f"\n{'='*80}")
logger.info(f"Total rows: {len(all_rows)}")

df = pd.DataFrame(all_rows)
df.to_csv(OUTPUT_FILE, index=False)
logger.info(f"‚úì Saved to {OUTPUT_FILE}")

print("\n" + "="*80)
print("üìä DATASET SUMMARY (2018 TEST)")
print("="*80)
print(f"\nShape: {df.shape}")
print(f"\nFirst 10 rows:")
print(df.head(10))
print(f"\n\nLast 10 rows:")
print(df.tail(10))

print(f"\n\nData types:")
print(df.dtypes)

print(f"\n\nMissing values:")
print(df.isnull().sum())

print(f"\n\nDNF distribution:")
print(df['dnf'].value_counts(dropna=False))

print(f"\n\nTotal drivers: {df['driver'].nunique()}")
print(f"Total events: {df['event'].nunique()}")

print(f"\n\nEvents in dataset:")
for event in df['event'].unique():
    count = len(df[df['event'] == event])
    print(f"  {event}: {count} rows")

print("\n" + "="*80)
print("‚úÖ TEST COMPLETE - Ready for full pipeline!")
print("="*80)
````

````python
# Regenerate feature snapshots (the parquet files seem corrupted)
import pandas as pd
import numpy as np
from pathlib import Path

# Load CSV data
df = pd.read_csv('f1_training_dataset_2018_2024.csv')
df['event_date'] = pd.to_datetime(df['event_date'], errors='coerce')
df = df.sort_values(['driver', 'event_date'])
print(f"Loaded {len(df)} rows from CSV")

# ===== DRIVER FEATURES =====
driver_features = []
for driver, group in df.groupby('driver'):
    group = group.sort_values('event_date')
    recent_5 = group.tail(5)['finishing_position'].mean()
    driver_features.append({
        'driver': driver,
        'TotalRaces': len(group),
        'RecentFormAvg': recent_5 if pd.notna(recent_5) else 10.0,
        'EloRating': group.iloc[-1].get('EloRating', 1500) if 'EloRating' in group.columns else 1500,
        'DriverExperienceScore': min(len(group) / 200, 1.0),
        'Wins': (group['finishing_position'] == 1).sum(),
        'Podiums': (group['finishing_position'] <= 3).sum()
    })

driver_df = pd.DataFrame(driver_features)
driver_df.to_parquet('models_spencer/driver_features_snapshot.parquet', index=False)
print(f"‚úì Driver features: {len(driver_df)} drivers")

# ===== CIRCUIT FEATURES =====
circuit_features = []
for (driver, circuit), group in df.groupby(['driver', 'circuit']):
    circuit_features.append({
        'driver': driver,
        'circuit': circuit,
        'CircuitAvgFinish': group['finishing_position'].mean(),
        'CircuitRaces': len(group),
        'CircuitBestFinish': group['finishing_position'].min()
    })

circuit_df = pd.DataFrame(circuit_features)
circuit_df.to_parquet('models_spencer/circuit_features_snapshot.parquet', index=False)
print(f"‚úì Circuit features: {len(circuit_df)} driver-circuit pairs")

# ===== TEAM FEATURES =====
team_features = []
for (year, team), group in df.groupby(['race_year', 'team']):
    team_features.append({
        'race_year': year,
        'team': team,
        'TeamAvgFinish': group['finishing_position'].mean(),
        'TeamPoints': group['points'].sum() if 'points' in group.columns else 0,
        'TeamPerfScore': 1 - (group['finishing_position'].mean() - 1) / 19  # Normalize 1-20 to 0-1
    })

team_df = pd.DataFrame(team_features)
team_df.to_parquet('models_spencer/team_features_snapshot.parquet', index=False)
print(f"‚úì Team features: {len(team_df)} team-year pairs")

print("\n‚úì All snapshots regenerated!")
````
